```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
df = pd.read_csv('/airbnb_london_clean.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df_num1 = df.select_dtypes('float64')
df_num1.head(10)
```

```python
df_mn_mx = pd.DataFrame()
for column in df_num1.columns:
    minimal = df_num1[column].min()
    maximum = df_num1[column].max()
    df_mn_mx[column + '_mn_mx'] = (df_num1[column] - minimal) / (maximum-minimal)
df_mn_mx.head(10)
```

```python
df_mn_mx.to_csv('airbnb_london_mn_mx.csv', index=False)
```

```python
df_num2 = df.select_dtypes('float64')
df_num2.head(10)
```

```python
df_avg = pd.DataFrame()
for column in df_num2.columns:
    avg = df_num2[column].mean()
    std = df_num2[column].std()
    df_avg[column + '_avg'] = (df_num1[column] - avg) / std
df_avg.head(10)
```

```python
df_avg.to_csv('airbnb_london_avg.csv', index=False)
```
