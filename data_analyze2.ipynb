```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airbnb_london_clean.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
colours = ['black', 'yellow']
sns.heatmap(df.isnull(), cmap=sns.color_palette(colours), cbar=False)
```

```python
avg_realSum = df['realSum'].mean()
print(f'  = {avg_realSum}')
```

```python
trimMean = stats.trim_mean(df['realSum'], 0.1)
print(f'   (10%) = {trimMean}')
```

```python
med = df['realSum'].median()
print(f' = {med}')
```

```python
variance = np.var(df['realSum'])
print(f' = {variance}')
```

```python
std_dev = df['realSum'].std()
print(f'  = {std_dev}')
```

```python
mad = np.mean(np.abs(df['realSum'] - avg_realSum))
print(f'   = {mad}')
```

```python
amed = np.median(np.abs(df['realSum'] - med))
print(f'A   = {amed}')
```
